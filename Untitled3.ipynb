```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from statsmodels.formula.api import ols
from statsmodels.tsa.stattools import kpss
from statsmodels.tsa.arima.model import ARIMA
from statsmodels.stats.outliers_influence import variance_inflation_factor
```

```python
Data = pd.read_excel(r'/content/gdrive/My Drive/crime data.xlsx')
Data.head()
```

Step 1 Create a multivariate time series; perform any interpolations.

I'm using the percent of respondent (%CrimeMuch) responded that more efforts should be done to crime and the estimation of homicides per 100k people (Homicide100K) in the US each year the GSS collects data.

```python
Data['year'] = pd.to_datetime(Data['year'])
Data.set_index('year', inplace=True)

# Interpolate missing values
Data_interpolated = Data.interpolate(method='linear')

# Rename columns to simplify usage
Data_interpolated.rename(columns={"Homicide100K": "Homicide100K", "%CrimeMuch": "%CrimeMuch"}, inplace=True)

# Create monthly dummies
Data_interpolated['year_Num'] = Data_interpolated.index.year
year_dummies = pd.get_dummies(Data_interpolated['year_Num'], prefix='year', drop_first=True)
Data_interpolated = pd.concat([Data_interpolated, year_dummies], axis=1)
```

Step 2 graph the relationship

```python
plt.figure(figsize=(14, 6))
plt.plot(Data_interpolated.index, Data_interpolated['%CrimeMuch'], label='% people thinking crime is too much (GSS)', color='blue')
plt.plot(Data_interpolated.index, Data_interpolated['Homicide100K'], label='Homicide Per 100K Population', color='orange')
x = ['year']
plt.title('Average Homicide Rates and Percent People Suggesting Increased Crackdown Of Crime')
plt.xlabel('time')
plt.ylabel('Values')
plt.legend()
plt.grid(True)
plt.show()
```

As we can see from the trends, while not always accurate, it seems that people's opinion on whether a lot of crimes happened this year has a loose relationship with the amount of homicide crimes commited this year.

```python
plt.figure(figsize=(8, 6))
plt.scatter(Data_interpolated['%CrimeMuch'], Data_interpolated['Homicide100K'], color='red', alpha=0.7)
plt.title('Homicide Per 100K Population vs. % people thinking crime is too much (GSS)')
plt.xlabel('Homicide Per 100K Population')
plt.ylabel('% people thinking crime is too much (GSS)')
plt.grid(True)
plt.show()
```

Step 3 Run a simple time series regression and interpret it.

```python
model_homicide_rate_only = ols('Q("%CrimeMuch") ~ Q("Homicide100K")', data=Data_interpolated).fit()
print(model_homicide_rate_only.summary())
```

Run a naive model predicting public opinion that crimes are too much and need to be cracked down more as a function of homicides, and we see, as we did in the scatterplot, that the results show a weak positive relationship, with homicide rates explaining about 30% of the variability in percent people thinking crimes are too much and need more forces to crack down.

Step 4 Run a time series regression with one X and trend. Interpret it. Perform autocorrelation diagnostics. Explain what you found.

We add a trend here.

```python
Data_interpolated['Trend'] = np.arange(1, len(Data_interpolated) + 1)
model_with_trend = ols('Q("%CrimeMuch") ~ Q("Homicide100K") + Trend', data=Data_interpolated).fit()
print(model_with_trend.summary())
```

Now we get a different regression result with trend.

Year dummies are useless in our case since years don't repeat, but we can try that to find out some outliers.

```python
Data_interpolated['Trend'] = np.arange(1, len(Data_interpolated) + 1)
model_with_trend = ols('Q("%CrimeMuch") ~ Q("Homicide100K") + Trend + ' + ' + '.join(year_dummies.columns), data=Data_interpolated).fit()
print(model_with_trend.summary())
```

In most cases the trends are smooth, but we found several outliers that in specific years such as 2021 when people suddenly change their opinions a lot about crime rates and efforts to crack down crime. It is therefore predicted that while in general people can correctly see the crime rates, specific events and public crisises like covid have great impacts on public views and make people think irrationally.

We have infinite standard errors here, representing that the "year dummies preditcions" are complete guesses without any data. This is expected since unlike months in the example, our year dummies has only one value each and we are basically prediciting with a dot.

Step 5 Consider running a time series regression with many Xs and trend. Interpret that. Check VIF.

We calculate the Variance Inflation Factor for the predictors.

```python
predictors = Data_interpolated[['Homicide100K', 'Trend']]

predictors_cleaned = predictors.select_dtypes(include=[np.number]).dropna()

vif_data = pd.DataFrame()
vif_data['Variable'] = predictors_cleaned.columns
vif_data['VIF'] = [variance_inflation_factor(predictors_cleaned.values, i) for i in range(predictors_cleaned.shape[1])]

print(vif_data)
```

High multicollinearity is observed. The predictors standard error may be noticeably inflated, the coefficients might not be reliable. This results proves sociologists' argument that general public don't have very good sense on crime rates.

Step 6 Run a first differenced time series regression. Interpret that.

```python
Data_interpolated['Diff_%CrimeMuch'] = Data_interpolated['%CrimeMuch'].diff()
Data_interpolated['Diff_Homicide100K'] = Data_interpolated['Homicide100K'].diff()
data_with_differences = pd.concat([Data_interpolated[['Diff_%CrimeMuch', 'Diff_Homicide100K']]], axis=1).dropna()
model_diff_with_change_dummies = ols('Q("Diff_%CrimeMuch") ~ Q("Diff_Homicide100K")',
                                     data=data_with_differences).fit()
print(model_diff_with_change_dummies.summary())
```

We can see that in this model our R-squared decreased significantly comparring to the two previous ones, sgguesting not much of variance is explained with our predictor. Durbin-Watson statistic is near to 2, indicating not much autocorrelation concern.

Step 7 Check your variables for unit roots. Do some tests. Interpret them.

The seventh step applies the KPSS test to assess the stationarity of the original and differenced series for public opinion on crime.

```python
def kpss_test(series):
    statistic, p_value, _, critical_values = kpss(series, regression='c')
    return {'KPSS Statistic': statistic, 'p-value': p_value, 'Critical Values': critical_values}

kpss_original = kpss_test(Data_interpolated['%CrimeMuch'].dropna())
kpss_diff = kpss_test(Data_interpolated['Diff_%CrimeMuch'].dropna())
print("KPSS Test Results (Original):", kpss_original)
print("KPSS Test Results (Differenced):", kpss_diff)
```

The test confirmed that both original and differenced series achieved stationarity.
