1. Run a naive OLS regression

I want to predict the "life expectancy at birth, total (years)" or "wdi_lifexp." Using describe to first look at it.

```python
df[['wdi_lifexp']].describe()
```

The next variable I will use is wdi_fertility, measured by the average total births per woman. I assume that countries with high fertility rates are mainly developing countries with less comprehensive pension funds and senior healthcare systems. On the other hand, the childbearing burden itself can lead to health issues and lower life expectancy.

```python
df[['wdi_fertility']].describe()
```

The following variable I will use is GDP per capita gle_cgdpc. I assume that GDP per capita can indicate a country's standard of living in some ways, so it will affect life expectancy.

```python
df[['gle_cgdpc']].describe()
```

I will predict life expectency at birth as a function of fertility rate and GDP per capita. I expect countries that have less fertility rates and more GDP per capita have higher life expectency at birth.

```python
gdp1 = smf.ols(formula = 'wdi_lifexp ~ wdi_fertility + gle_cgdpc', data = df).fit()
print (gdp1.summary())
```

The results meet our expectations. For one more baby an average woman in a country gives birth to, life expectency at birth is going to decrease for 4.28 years. For one more dollar a citizen of a country produces on average in a year, the life expectancy at birth will increase for 0.0002.

These two variables allow us to predict life expectency at birth with 74.1% more accuracy, compared to just guessing the average life expectency.

2. Run a first differences regression.

Install first difference model.

```python
!pip install linearmodels

from linearmodels.panel import FirstDifferenceOLS
```

```python
columns = ['ccode', 'year', 'wdi_lifexp', 'wdi_fertility', 'gle_cgdpc']
df1 = df[columns].dropna()
```

```python
# Set the MultiIndex for panel data
df1 = df1.set_index(['ccode', 'year'])

# Define the dependent and independent variables
y = df1['wdi_lifexp']
X = df1[['wdi_fertility', 'gle_cgdpc']]

# Fit the first-differenced panel data model
fdmodel = FirstDifferenceOLS(y, X)
results = fdmodel.fit(cov_type='clustered', cluster_entity=True)

print(results)
```

By using first difference estimation, we can see that for one more baby a woman gives birth to on average in a country, life expectency at birth is expected to decrease for 1.0044 years. On the other hand, for one more dollar produced in GDP per capita of a country, life expectency is expected to increase for 1.739e-05 years. Coefficients are much smaller in magnitude.

We can see that the fertility rate still significantly negatively impacts life expectancy at birth, but GDP per capita only has a small impact and is not statistically significant, with a p-value higher than 0.05. Fertility rates appear more related to life expectancy at birth.
