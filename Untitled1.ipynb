```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import statsmodels.api as sm
import statsmodels.formula.api as smf
url = 'https://www.qogdata.pol.gu.se/data/qog_bas_cs_jan24.xlsx'
df = pd.read_excel(url)
df.head()
```

```python
from sklearn.linear_model import LinearRegression, Lasso, Ridge
from sklearn.neighbors import KNeighborsRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split
from sklearn.model_selection import cross_val_score, GridSearchCV
from sklearn.metrics import mean_squared_error
```

1. Create training and testing datasets:
I'm still predicting "life expectancy at birth, total (years)" or "wdi_lifexp." Variables rleating to society and quality of life of citizens will be used to predict the life expectancy at birth.

Society: The Bayesian Corruption Index 0-100 (bci_bci,) Military Expenditure % of GDP (wdi_expmil,) Arable land % of total land (wdi_araland,) political stability (wbgi_pve.)

Quality of Life: Population density/people per square kilometer (wdi_popden,) unemployment rate (wdi_unempilo,) GDP per capita 2011 dollar (mad_gdppc,) Electricity indicator 0-100 (wef_elec.)

```python
df_cleaned = df.dropna(subset=['wdi_lifexp', 'wdi_expmil', 'bci_bci', 'wdi_araland', 'wbgi_pve', 'wdi_popden', 'wdi_unempilo', 'mad_gdppc', 'wef_elec'])
X = df_cleaned[['wdi_expmil', 'bci_bci', 'wdi_araland', 'wbgi_pve', 'wdi_popden', 'wdi_unempilo', 'mad_gdppc', 'wef_elec']]
y = df_cleaned['wdi_lifexp']
```

2. Run ols regression and get the main squared error

Using the hold out method, typically 70%-30%, so we will use 30% as well

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.3,random_state=42)
```

```python
ols_model = LinearRegression()
ols_model.fit(X_train, y_train)
y_pred = ols_model.predict(X_test)

print('Mean Squared Error:', np.sqrt(mean_squared_error(y_test, y_pred)))
rmse = np.sqrt(np.sqrt(mean_squared_error(y_test, y_pred)))
print('RMSE:', rmse)
```

```python
from sklearn.metrics import mean_squared_error, r2_score
r_squared = r2_score(y_test, y_pred)
print('R-squared:', r_squared)
```

At a first glance it seems that our model has a pretty big mean squared error (4.24) and isn't expected to have the capability of predicting the life expectancy at birth very accurately.

However, since our outcome variable, life expectancy at birth, measured by years are pretty big in numbers in general, it is also possible that our model remains capable to predict a rough number of life expectancy at birth with a root mean squared error 2.06.


3. Interpret the coefficients:

```python
coefficients = pd.Series(ols_model.coef_, index=X.columns)
coefficients
```

wdi_expmil: 1% of increase in GDP% spend on military is associated with a small decrease of 	-0.178868 years in life expectancy at birth.
wef_elec: 1 unit of increase in electricity on 1 1-100 scale is associated with a small increase of 0.293842 years in life expectancy at birth.
bci_bci: 1 unit of increase in the Bayesian Corruption Index for the country is associated with a small decrease of 0.098231 years in life expectancy at birth.

4. Run a random forest model over the same data. Select the best parameters using grid-search.

```python
rf1 = RandomForestRegressor()

n_estimators = [50, 100, 300]
max_depth = [None, 1, 5]
min_samples_split = [2, 5, 10]
min_samples_leaf = [1, 2, 4]

hyperF = dict(n_estimators = n_estimators, max_depth = max_depth,
              min_samples_split = min_samples_split, min_samples_leaf = min_samples_leaf)

gridF = GridSearchCV(rf1, hyperF, cv = 3, verbose = 1,
                      n_jobs = -1)
bestF = gridF.fit(X_train, y_train)

print("Best Parameters: %f using %s" % (bestF.best_score_, bestF.best_params_))
```

5. Run the best model and get the mean-squared-error on your testing data.

```python
rf_best = RandomForestRegressor(max_depth= 5, min_samples_split= 10, n_estimators= 100, min_samples_leaf=4)
rf_best.fit(X_train, y_train)
y_pred = rf_best.predict(X_test)

print('Mean Squared Error:', np.sqrt(mean_squared_error(y_test, y_pred)))
```

Mean squared error from random forest regressor is 3.72, which is smaller than 4.24 of ols regression. This suggests that the random forest model performs better than than ols model in terms of accuracy in predicting.

```python
r_squared = rf_best.score(X_test, y_test)
print('R-squared:', r_squared)
```

The r squared also increased a bit comparring to the ols regression model, indicating that more proportion of variance can be explained with our new model.

6. Interpret fature importance.

```python
import numpy as np
import matplotlib.pyplot as plt

importance = rf_best.feature_importances_
indices = np.argsort(importance)[::-1]
feature_names = X.columns

f, ax = plt.subplots(figsize=(10, 8))
plt.title("Feature ranking", fontsize=20)
plt.bar(range(X.shape[1]), importance[indices], color="b", align="center")
plt.xticks(range(X.shape[1]), feature_names[indices], rotation=90)  # Use indices to sort feature names
plt.xlim([-1, X.shape[1]])
plt.ylabel("Importance", fontsize=18)
plt.xlabel("Features", fontsize=18)
plt.show()
```

From the plot rank we can see that electricity is the dominantly important variable, which suggests that when including electrity to predict life expectancy at birth, the number will be a lot more accurate.
