```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import statsmodels.api as sm
import statsmodels.formula.api as smf
import scipy.stats as stats
!pip install requests
import requests
import io
from tqdm import tqdm
import zipfile
```

```python
url = 'https://gss.norc.org/content/dam/gss/get-the-data/documents/stata/GSS_stata.zip'

response = requests.get(url, stream=True)
total_size = int(response.headers.get('content-length', 0))
block_size = 1024

tqdm_bar = tqdm(total=total_size, unit='iB', unit_scale=True)
content = io.BytesIO()

for data in response.iter_content(block_size):
    tqdm_bar.update(len(data))
    content.write(data)

tqdm_bar.close()

if total_size != 0 and tqdm_bar.n != total_size:
    print("Error in downloading the file")
else:
    print("Download completed")
```

1. Run a multiple linear probability model (have at least 2 Xs in the model). Tell me how you think your independent variables will affect your dependent variable. Interpret your results. Were your expectations correct? Why or why not?

My variables:

agekdbrn: the number of age of respondent when their first kid was born.

educ: the highest of years of school the respondent completed

paeduc: the highest of years of school the father of the respondent completed

polviews: the political view of the respondent measured on a 1-7 scale, extremely liberal, liberal, slightly liberal, moderate, slightly conservative, conservative, extremely conservative.

homosex: "the level of wrong" the respondent think sexual relationship between two people of the same gender is, measured on a 1-4 scale. 1= always wrong, 2= almost always wrong, 3= sometimes wrong, 4=not wrong at all.

```python
with zipfile.ZipFile(content) as z:
    file_list = z.namelist()
    stata_files = [file for file in file_list if file.endswith('.dta')]

    if stata_files:
        stata_file = stata_files[0]

        with z.open(stata_file) as stata_file_stream:
          columns_to_load = ['year', 'agekdbrn', 'educ', 'paeduc', 'polviews', 'homosex']
          print("Loading selected columns from Stata file with numeric labels")
          df_numeric = pd.read_stata(stata_file_stream, columns=columns_to_load, convert_categoricals=False)
          print("Data with numeric labels loaded successfully")

        with z.open(stata_file) as stata_file_stream:
          print("Loading selected columns from Stata file with string (categorical) labels")
          df_categorical = pd.read_stata(stata_file_stream, columns=columns_to_load)
          print("Data with categorical labels loaded successfully")
```

Get rid of inapplicable or skipped values:

```python
df_numeric = df_numeric[~df_numeric['year'].isin([-100, -99, -98, -97])]
df_numeric = df_numeric[~df_numeric['agekdbrn'].isin([-100, -99, -98, -97])]
df_numeric = df_numeric[~df_numeric['polviews'].isin([-100, -99, -98, -97])]
df_numeric = df_numeric[~df_numeric['educ'].isin([-100, -99, -98, -97])]
df_numeric = df_numeric[~df_numeric['paeduc'].isin([-100, -99, -98, -97])]
df_numeric = df_numeric[~df_numeric['homosex'].isin([-100, -99, -98, -97])]
```

homosex will be my dependent variable. i assume that people who are more liberal, more educated, have more educated father, and have their first kid at older age (which means they are relatively elite) will be more open to homosexual relationships and tend to think that sex between two people of the same gender is not wrong at all.

homosex is the variable which represent's opinion on whether homosexual relationship is wrong, 4= always not worng at all, 1, 2, 3= different levels of belief that homosexual relationship is wrong. I will recode homosex into a binary variable.

```python
conditions = [(df_numeric['homosex'] < 4) , (df_numeric['homosex'] > 0)]
choices = [0, 1]
df_numeric['homosex_notwrong'] = np.select(conditions, choices, default=np.nan)
```

```python
pd.crosstab(index=df_numeric["homosex_notwrong"], columns="count")
```

```python
lm1 = smf.ols(formula = 'homosex_notwrong ~ polviews + educ + paeduc + agekdbrn', data = df_numeric).fit()
print (lm1.summary())
```

poltical view (polviews) is negatively associated with the belief that homosex is not wrong, which represents more conservative people believe less that homosex is not wrong at all, and the association is statistically significant.

Education (Educ) is positively associated with the belief that homosex is not wrong, which suggests that more educated people tend to believe that homosex is not wrong, the association is statistically significant.

father's education (paeduc) is also positively associated with the belief that homosex is not wrong, and the association is statistically significant.

the age when the first kid was born (agekdbrn) is postively associated with the belief that homosex is not wrong, and the association is statistically significant. People have kids later in their life are expected to believe more that homosexual relationship is not wrong at all.

The result matches our expectation that more liberal, more educated and more elite people are less likely to think that sex between people of the same gender is wrong.

2. Run a multiple (binary) logistic model. (It can be the same as the above LPM or a new model.) If it is a new model, tell me how you think your independent variables will affect your dependent variable. Interpret your results in the logit scale. Were your expectations correct? Why or why not?

```python
logit1 = sm.formula.logit(formula = 'homosex_notwrong ~ polviews + educ + paeduc + agekdbrn', data = df_numeric).fit()
print (logit1.summary())
```

Logit Model shows the same results. More liberal people, more educated people, people with a more educated father, and people have their first kid relatively late tend to believe that homosexual relationship is always not wrong. All associations are statistically significant.

4. Get predicted probabilities from your logit model in Question 2 for some constellations of X values and interpret the results.

```python
def logit2prob (logit):
    odds = np.exp(logit)
    prob = odds / (1 + odds)
    return(prob);

intercept = logit1.params.Intercept
b_polviews = logit1.params.polviews
b_educ = logit1.params.educ
b_paeduc = logit1.params.paeduc
b_agekdbrn = logit1.params.agekdbrn
```

```python
logits_exh = intercept + (1 * b_polviews) + (40 * b_agekdbrn) + (20 * b_educ) + (20 * b_paeduc)
logit2prob(logits_exh)
```
